```python
def encode_categorical_columns(df):

    df_encoded = df.copy()  # we must  Make a copy to avoid modifying the original DataFrame

    for col in df.columns:
        # check if the column is non-numeric or not?! :
        if not pd.api.types.is_numeric_dtype(df[col]):
            uv = df[col].nunique()  # Count unique values
            #uv: number of unnique values
            if uv < 13: #one hot encoding
                df_encoded = pd.get_dummies(df_encoded,drop_first=True, columns=[col], prefix=col)
            elif(13<=uv<=25):
            #  frequency Encoding for columns with more than 5 unique values
                freq_encoding = df[col].value_counts() / len(df)
                df_encoded[col] = df[col].map(freq_encoding)

    return df_encoded
```

```python
def remove_outliers_zscore(df, threshold=1.6):

    numeric_cols = df.select_dtypes(include=[np.number]).columns       # we must Select only numeric columns for the calculation
    
    dff = pd.DataFrame()
    for col in numeric_cols:
        col_mean = df[col].mean()
        col_std = df[col].std()
        dff[col] = np.abs((df[col] - col_mean) / col_std)
    
    dfff = df[(dff < threshold).all(axis=1)]
    
    return dfff

df_cleaned = remove_outliers_zscore(df)
```

```python
class CustomStackedRegressor:
    def __init__(self, estimators, final_estimator):

        #estimators: List of tuples with the format (name, model) for base models
        #final_estimator: meta-model to train on base model predictions

        self.base_models = estimators
        self.meta_model = final_estimator
        self.fbm = []     #fbm : fitted base models

    def fit(self, X, y): #this func train the base models and then use their predictions to train the meta-model
        #X: training features
        #y: target values

        self.fbm = []                # store the fitted base models
        bp = []                  # bp : base predictions 
        #train each base model on the training data
        for name, model in self.base_models:     # base_models : [(name1,model1),(...,...),(...,...)]
            fitted_model = model.fit(X, y)
            self.fbm.append(fitted_model)
            bp.append(fitted_model.predict(X))
        
        meta_features = np.column_stack(bp) # stack predictions from each base model as new features for meta-model
        self.meta_model.fit(meta_features, y)          # train meta-model on these new features

    def predictt(self, X):  #this func make predictions with the stacked model by first getting predictions from base models and then feeding them into the meta-model
        
        #X: Test features to predict target values.
        
        #what returns?
        #predictions of the stacked model :)
       ############################################# 
        #generate predictions for each base model:
        bp = []
        for model in self.fbm:
            prediction = model.predict(X)
            bp.append(prediction)

        
        #  consider predictions from each base model as new features for the meta-model:
        meta_features = np.column_stack(bp)
        
        return self.meta_model.predict(meta_features)      # use the meta-model to predict the final output(last step)
    
    #very hard!!!!!!! but finished
```

```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, ElasticNet,LinearRegression
from sklearn.kernel_ridge import KernelRidge
from sklearn.ensemble import GradientBoostingRegressor
#, StackingRegressor
from sklearn.metrics import mean_squared_error, r2_score

#X_training  =data.drop(['Heating', 'RoofMatl', 'GarageQual', 'Electrical', 'Utilities_NoSeWa','SalePrice'] ,axis=1)   #this columns are absent in df_test
X_training  =data.drop(['SalePrice'] ,axis=1)
y_training = data['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X_training, y_training, test_size=0.2, random_state=42)

# Define the base models for stacking
base_models = [
    ('lasso', Lasso(alpha=0.1)),  # Lasso Regression
    ('elastic_net', ElasticNet(alpha=0.1, l1_ratio=0.5)),  # Elastic Net Regression
    ('kernel_ridge', KernelRidge(alpha=1.0, kernel='linear')),  # Kernel Ridge Regression
    ('gradient_boosting', GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3))  # Gradient Boosting
]

# Define the meta-model
meta_model = LinearRegression()
stacked_regressor = CustomStackedRegressor(estimators=base_models, final_estimator=meta_model)  # Create the Stacked Regressor
stacked_regressor.fit(X_train, y_train)

y_pred = stacked_regressor.predictt(X_test)     #for training dataset
y_pred_test = stacked_regressor.predictt(TEST_X)   #for dataset test
##################################################################3
# Calculate Eror
def calculate_r2(y_true, y_pred):
    tss = np.sum((y_true - np.mean(y_true)) ** 2)   #total sum of square
    rss = np.sum((y_true - y_pred) ** 2)           #3residual sum of square
    r2_score = 1 - (rss / tss)                       # calculate r^2
    return r2_score

def calculate_rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

rmse = calculate_rmse(y_test, y_pred)
r2 = calculate_r2(y_test, y_pred)
################
test_rmse = calculate_rmse(TEST_y, y_pred_test)   # y is saleprice column of df_test
test_r2 = calculate_r2(TEST_y, y_pred_test)
######################################################################
print(f"Root Mean Squared Error: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")
print("###################################### \t For data set test :")
print(f"Root Mean Squared Error: {test_rmse:.4f}")
print(f"R² Score: {test_r2:.4f}")
```
